# LDA

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("Historia_Climatica.csv")
X = df[["Temperatura_C", "Humedad_%", "Vel_Viento_mps", "Presion_hPa"]]

# Escalar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Etiquetas falsas para prueba (reemplazar con reales si existen)
import numpy as np
y_fake = np.random.randint(0, 3, size=X.shape[0])

lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_scaled, y_fake)

# Visualización
plt.figure(figsize=(8,5))
sns.scatterplot(x=X_lda[:, 0], y=X_lda[:, 1], hue=y_fake, palette='tab10')
plt.title("Reducción de Dimensionalidad con LDA (datos simulados)")
plt.show()
```
