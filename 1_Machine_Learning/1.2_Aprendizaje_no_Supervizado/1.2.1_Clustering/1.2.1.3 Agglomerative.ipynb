# Agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar los datos
df = pd.read_csv('Historia_Climatica.csv')
print(df.head())

# Seleccionar variables numéricas para clustering
X = df[["Temperatura_C", "Humedad_%", "Vel_Viento_mps", "Presion_hPa"]]

# Ajustar el modelo Agglomerative
modelo = AgglomerativeClustering(n_clusters=4)
y_pred = modelo.fit_predict(X)

# Agregar etiquetas de clúster al DataFrame original
df["Cluster"] = y_pred
```

```python
# Visualizar resultados con un par de variables
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="Temperatura_C", y="Humedad_%", hue="Cluster", palette="tab10")
plt.title("Clustering Jerárquico: Temperatura vs Humedad")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Humedad (%)")
plt.tight_layout()
plt.show()
```
